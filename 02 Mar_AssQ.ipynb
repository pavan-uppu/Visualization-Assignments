Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
# Scatter plot is a type od data visualization that is used to display patterns or correlation between data points
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
# subplot() function in matplotlib is used to create multiple plots within the same figure.
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Line 4")

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

```python
# bar plot is a type of data visualization that represents data as rectangular bars. It is commonly used to display and compare categorical data

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(8, 4))
plt.bar(company, profit, color='#d4e069')
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Vertical Bar Plot")
plt.show()

plt.figure(figsize=(8, 4))
plt.barh(company, profit, color='#808458')
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Horizontal Bar Plot")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

```python
# box plot is a graphical representation of a dataset's summary statistics, displaying information about
# the distribution, spread, and potential outliers of the data

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=False, labels=['Box 1', 'Box 2'])
plt.xlabel("Value")
plt.title("Box Plot of Box 1 and Box 2")
plt.show()
```
